```python
import numpy as np
import matplotlib.pyplot as plt

# Use RAW string (r"") and forward slashes (/) - this works on Windows
# Your confirmed file
filename = r"D:\IIT\L6\FYP\ChagaSight\data\processed\2d_images\sami_trop\24028_img.npy"
img = np.load(filename)
print("Shape:", img.shape)    # (3, 24, 2048)
print("Dtype:", img.dtype)    # float32

# Plot the 3 contours
fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].imshow(img[0], cmap='gray')
axs[0].set_title('RA Contour')
axs[1].imshow(img[1], cmap='gray')
axs[1].set_title('LA Contour')
axs[2].imshow(img[2], cmap='gray')
axs[2].set_title('LL Contour')
plt.tight_layout()
plt.show()
```
